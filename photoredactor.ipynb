```python
# Вырезание, вставка и объединение изображений

from PIL import Image
pil_im = Image.open("../IMG_6325.jpg")

box = (100,100,400,400)
region = pil_im.crop(box)
#region.show()
region = region.transpose(Image.ROTATE_180)
#region.show()

pil_im.paste(region, box)

#pil.save('Image_rotated.jpg')
show()
```

```python
# Изменение размера и поворот

from PIL import Image
image = Image.open("../IMG_6325.jpg")

out = image.resize((128,128))

out = image.rotate(45)

out.show()
```
